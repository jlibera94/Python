    "def filter1(x, threshold):\n",
    "    # Your answer here\n",
    "    y = []\n",
    "    for xi in x:\n",
    "        if xi > threshold:\n",
    "            y.append(xi)\n",
    "    return y\n",
    "\n",
    "def filter2(x, threshold):\n",
    "    # Your answer here\n",
    "    return [xi for xi in x if xi > threshold]\n",
    "\n",
    "def filter3(x, threshold):\n",
    "    # Your answer here\n",
    "    return filter(lambda xi: xi > threshold, x)"
   ]
